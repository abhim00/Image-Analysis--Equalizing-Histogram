```python
#https://gist.github.com/magsol/6628071

%pylab inline
import scipy.ndimage as nd
import matplotlib.cm as cm


img = nd.imread("mammogram.jpg", flatten=True)
axis('off')
imshow(img, cmap=cm.gray)
```

```python
#Flatten argument forces the image to be gray scale.
#Assuming the pixel intensities derive from unsigned 8-bit integers,
#the values in the range are [0,256)

hist(img.flatten(), bins=256)
show()
```

```python
#CDF stands for the cumulative distributive function. 
#For each x, it is the sum of all the values before it.
#Going back to the histogram, this equates with each pixel intensity [0,256) with the sum of all the hist values before it.

histogram, edges = np.histogram(img.flatten(), bins=256)
cdf = histogram.cumsum().astype("float")
xlim([0,255])
#print(cdf)
plot(cdf)
```

```python
cdf_scaled = np.round(((cdf-cdf.min())/(img.shape[0]*img.shape[1]-cdf.min()))*(256-1))
xlim([0,255])
plot(cdf_scaled)
```

```python
img2 = cdf_scaled[img.astype("uint8")]
axis('off')
f, (ax1,ax2) = subplots(1,2)
ax1.set_title("Original")
ax1.axis('off')
ax1.imshow(img, cmap = cm.gray)
ax2.set_title("Equalized")
ax2.axis('off')
ax2.imshow(img2, cmap = cm.gray)
```

```python
hist2, edges2 = np.histogram(img2.flatten(), bins=256)
cdf2 = hist2.cumsum().astype('float')
xlim([0,255])
plot(cdf2)
```
